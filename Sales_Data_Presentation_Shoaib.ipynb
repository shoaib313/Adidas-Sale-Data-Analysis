```python
from collections import Counter
def descriptive_statistics_and_plot(df, column):
    mean = df[column].mean()
    median = df[column].median()
    counter = Counter(df[column])
    print(counter.most_common())
    mode = counter.most_common(1)[0][0]
    range_value = df[column].max() - df[column].min()
    variance = df[column].var()
    std_deviation = df[column].std()
    
    print(f"Descriptive Statistics for {column}:")
    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode}")
    print(f"Range: {range_value}")
    print(f"Variance: {variance}")
    print(f"Standard Deviation: {std_deviation}")
    print("\n")
    
    # Plot histogram
    plt.figure(figsize=(8, 5))
    sns.histplot(df[column], kde=True, color='k')
    plt.axvline(mean, color='r', linestyle='--', label=f'Mean: {mean:.2f}')
    plt.axvline(median, color='g', linestyle='-', label=f'Median: {median:.2f}')
    plt.axvline(mode, color='b', linestyle='-', label=f'Mode: {mode:.2f}')
    plt.legend()
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
descriptive_statistics_and_plot(data,'Price per Unit')
descriptive_statistics_and_plot(data,'Units Sold')
descriptive_statistics_and_plot(data,'Total Revenue')
descriptive_statistics_and_plot(data,'Operating Profit')
```

## 1. Cleaning and Handling Missing Data
Data cleaning involves handling missing values to ensure data quality. Missing values can be handled by removing or imputing them.


```python
print(data.isnull().sum())
```

```python


def replace_missing_with_mean(df, numerical_cols):
    imputer = SimpleImputer(strategy='mean')
    df[numerical_cols] = imputer.fit_transform(df[numerical_cols])
    return df
def replace_missing_with_median(df, numerical_cols):
    imputer = SimpleImputer(strategy='median')
    df[numerical_cols] = imputer.fit_transform(df[numerical_cols])
    return df
numerical_cols_mean = ['Price per Unit']
numerical_cols_median = ['Total Revenue','Units Sold', 'Operating Profit']

df_clean = replace_missing_with_mean(data, numerical_cols_mean)
df_clean = replace_missing_with_median(data, numerical_cols_median)
df_clean.isnull().sum()
```

```python
def remove_outliers(df, cols):
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[((df[cols] < (Q1 - 1.5 * IQR)) | (df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
    df_clean = df[~((df[cols] < (Q1 - 1.5 * IQR)) | (df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df_clean, outliers
numerical_cols = ['Units Sold', 'Price per Unit', 'Total Revenue', 'Operating Profit']

# Remove outliers
df_clean_n, outliers= remove_outliers(df_clean, numerical_cols)
print(len(df_clean_n))
#sns.pairplot(df_clean_n)
plt.figure(figsize=(7, 5))
plt.boxplot([df_clean_n['Total Revenue'],  df_clean_n['Operating Profit']], labels=[ 'Total Revenue', 'Operating Profit'], showmeans=True)
plt.title('Distribution of Profit, Revenue, Units Sold, and Price per Unit')
plt.ylabel('($)')
plt.show()
plt.figure(figsize=(7, 5))
plt.boxplot([df_clean_n['Units Sold'],  df_clean_n['Price per Unit']], labels=[ 'Units Sold', 'Price per Unit($)'], showmeans=True)
plt.title('Distribution of Units Sold, and Price per Unit')
plt.ylabel('(Value)')
plt.show()
```
